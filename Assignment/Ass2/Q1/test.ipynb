```python
import random

def generate_sparse_matrix(n, m):
    max_non_zero_elements = 5 * 10**5
    total_non_zero = 0
    
    matrix = []
    
    # Generate each row
    for i in range(n):
        # Randomly decide how many non-zero elements in this row (max: m)
        c = random.randint(0, m)
        total_non_zero += c
        
        # Ensure total non-zero elements don't exceed the limit
        if total_non_zero > max_non_zero_elements:
            # If we're about to exceed the limit, adjust the number of non-zero elements for the last row
            c = max_non_zero_elements - (total_non_zero - c)
            total_non_zero = max_non_zero_elements
        
        # Randomly select the column indices for non-zero elements (ensure they are sorted)
        indices = random.sample(range(m), c)
        indices.sort()
        
        row_data = [c]  # First element is the number of non-zero elements in this row
        
        # For each index, randomly select a value between -100 and 100
        for idx in indices:
            value = random.randint(-100, 100)
            row_data.append(idx)
            row_data.append(value)
        
        matrix.append(row_data)
        
        if total_non_zero == max_non_zero_elements:
            break  # Stop if we have reached the maximum non-zero elements
    
    # Print matrix in the required format
    for row in matrix:
        print(" ".join(map(str, row)))
    print()  # Blank line after matrix

# Example usage:
n = 100 # Number of rows
m = 100  # Number of columns
print(n, m)
generate_sparse_matrix(n, m)
print(n, m)
generate_sparse_matrix(n, m)
```
